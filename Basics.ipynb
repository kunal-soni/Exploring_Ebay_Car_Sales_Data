Project to explore Ebay Car Sales Data. The data contains car details. Aim of the project is to clean the data and analyze the included used car listings.

```python
# Edit column names and rename few to snake case
autos = autos.rename(columns = {"yearOfRegistration" : "registration_year",
                       "monthOfRegistration":"registration_month",
                       "notRepairedDamage":"unrepaired_damage",
                       "dateCreated":"ad_created"})
autos.head()
```

```python
# Basic data exploration
autos.describe()
```

```python
# Describe including non-numeric columns (categorical)
autos.describe(include = 'all')
```

```python
# Convert price and odometer columns to numeric
autos['odometer'] = autos['odometer'].str.replace("km","")
autos['odometer'] = autos['odometer'].str.replace(",","")
autos['odometer'] = autos['odometer'].astype(int)

autos['price'] = autos['price'].str.replace("$","")
autos['price'] = autos['price'].str.replace(",","")
autos['price'] = autos['price'].astype(int)

autos.tail()
```

```python
# Rename odometer column
autos = autos.rename(columns = {"odometer":"odometer_km"})
autos.head(2)
```

```python
# Explore data for odometer_km column
print(autos['odometer_km'].unique().shape)
print(autos['odometer_km'].value_counts())
print(autos['odometer_km'].describe())
```

# Auto brands
We have 40 unique auto brands in the dataset. 3 top brands are listed and have > 10% of the total. 17 brands are above 1%, 23 brands are below 1%.

For our analysis we selected the top 20 brands as they appear to be more relevant.

```python
# Dictionary of aggregate data for brands
brands_mean_prices = {}

for brand in top20_brands:
    current_brand = autos[autos["brand"] == brand]
    mean_price = current_brand["price"].mean()
    brands_mean_prices[brand] = int(mean_price)

brands_mean_prices
```

# Top 20 brands mean prices
We see that Sonstige Autos has maximum mean price of 12338. The minimum is Renault: 2474. Audi, BMW, Mini, Mercedes Benz are the higher end prices, as the mean price is above 8000

# Exploring mileage


```python
brands_mean_mileage = {}
# Use loop to calculate mean mileage for the brands
for brand in top20_brands:
    current_brand = autos[autos["brand"] == brand]
    mean_mileage = current_brand["odometer_km"].mean()
    brands_mean_mileage[brand] = int(mean_mileage)

# Convert both dictionaries to series objects
mean_mileage = pd.Series(brands_mean_mileage).sort_values(ascending=False)
mean_prices = pd.Series(brands_mean_prices).sort_values(ascending=False)
print(mean_mileage)
print(mean_prices)
```

```python
topbrandsDF = pd.DataFrame(mean_mileage, columns = ['mean_mileage'])
topbrandsDF['mean_price'] = mean_prices
topbrandsDF
```

# Top 20 brands mileage and price analysis
From above dataframe, we see there is a slight increase in price as mileage is going down. But it is not consistent, there are 3 brands in top 5 that have higher price - BMW, Mercedes and Audi.

```python
# Split odometer_km into groups 
autos['odometer_km'].value_counts()
```

```python
# Break into bins
bins = [0,40000,70000,100000,130000,160000]

# Create the names for the bins.
# Group names will have 1 less element in list than the bins defined
group_names = ["<40K","40K-70K","70K-100K","100K-130K","130K+"]

# For purchase_data, a new column of Age Group is added
autos["odometer_group"] = pd.cut(autos["odometer_km"],bins,labels=group_names)
autos.head()
```

```python
# Use aggregation to see if average prices follow any patterns based on mileage
autos.groupby('odometer_group')['price'].mean().sort_values(ascending=False)
```

# Analysis of price compared to odometer group
It is seen that as the odometer mileage increases, price decreases and they are inversely proportional.

For Odomter reading < 40K, mean price is 15213. 70K-100K has 8670, 130K+ has 3768.
